```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import pandas as pd
file_path = 'quotes.csv'
data = pd.read_csv(file_path)
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
data.nunique()
```

```python
data.isna().sum()
```

```python
data.dropna(inplace=True)
```

```python
data.info()
```

```python
quotes = " ".join(data['Quote'])
wordcloud = WordCloud(width=800, height=400, background_color="white").generate(quotes)
```

```python
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```

TOP 10 MOST QUOTED AUTHORS

```python
tag_counts = data['Author'].str.split(',').explode().value_counts()

plt.figure(figsize=(10, 6))
tag_counts.head(20).plot(kind='barh', color='lightgreen')
plt.title("Top 10 MOST QUOTED AUTHORS")
plt.xlabel("Frequency Of Quotes")
plt.ylabel("NAME OF AUTHORS")
# plt.gca().invert_yaxis()
plt.show()
```

TOP 10 MOST COMMON TAGS

```python
tag_counts = data['Tags'].str.split(',').explode().value_counts()

plt.figure(figsize=(10, 6))
tag_counts.head(20).plot(kind='barh', color='lightgreen')
plt.title("Top 10 MOST COMMON TAGS")
plt.xlabel("Frequency")
plt.ylabel("TAGS")
plt.gca().invert_yaxis()
plt.show()
```
